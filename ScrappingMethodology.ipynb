```python
from urllib.request import urlopen
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
def cities_scrapping():
    url = 'https://www.tripadvisor.fr/Restaurants-g187144-Ile_de_France.html'
    raw_data = urlopen(url).read()
    soup = bs(raw_data, 'lxml')
    pages = soup.findAll('a', {'class' : 'pageNum taLnk'})[-1]['data-page-number']
    pages = int(pages)
    pages = [page for page in range(1000000) if page%2 == 0][:pages]
    result = pd.DataFrame()
    
    for page in pages:
        if page < 2:
            raw_data = urlopen(url).read()
            soup = bs(raw_data, 'lxml')
            soup = soup.findAll('div', {'class' : 'geo_name'})
        else:
            url = "https://www.tripadvisor.fr/Restaurants-g187144-oa{}0-Ile_de_France.html".format(page)
            raw_data = urlopen(url).read()
            soup = bs(raw_data, 'lxml')
            soup = soup.find('ul', {'class' : 'geoList'})
            soup = soup.findAll('li')

        names = [cities.text.strip() for cities in soup]
        urls = [cities.a['href'] for cities in soup]
        new_dataframe = pd.DataFrame({'name' : names, 'url' : urls})
        result = pd.concat([result, new_dataframe])
        
    result = result.reset_index(drop=True)
    result['name'] = result['name'].str.replace(' : restaurants', '')
    result['name'] = result['name'].str.replace('\n', '')
    result['url'] = 'https://www.tripadvisor.fr' + result['url']
    return result


def restaurants_scapping(url):
    raw_data = urlopen(url).read()
    soup = bs(raw_data, 'lxml')
    pages = soup.findAll('a', {'class' : 'pageNum taLnk'})[-1]['data-page-number']

    raw_data = urlopen(url).read()
    soup = bs(raw_data, 'lxml')
    soup = soup.findAll('a', {'class' : 'property_title'})
    names = [restaurants.text.strip() for restaurants in soup]
    urls = [urls['href'] for urls in soup]
    result = pd.DataFrame({'name' : names, 'url' : urls})
    result['url'] = 'https://www.tripadvisor.fr' + result['url']
    return pages
    
```

```python
result = cities_scrapping()
result.to_csv('cities.csv')
```
